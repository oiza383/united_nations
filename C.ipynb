```python
%%sql

--Find the maximum, minimum, and average freight from the 'Orders' table.

SELECT *

FROM Orders;
```

```python
%%sql

SELECT 
    MAX(Freight) AS max_freight,
    MIN(Freight) AS min_freight,
    AVG(Freight) AS avg_freight
FROM Orders;
```

```python
%%sql
--How many unique customers have made orders? Use the 'Orders' table.
SELECT 
    COUNT(DISTINCT CustomerID) 
    
FROM Orders;
```

```python
%%sql
/*What is the total price for each order? You need to multiply the unit price by the quantity and 
subtract the discount for each order detail, then sum all the order details for each order.
Use the 'OrderDetails' table*/

SELECT
    OrderID,
    SUM((UnitPrice * Quantity)*(1-Discount)) AS each_price
FROM OrderDetails

GROUP BY OrderID;
```

```python
%%sql

--What are the 5 smallest orders processed
SELECT
    OrderID,
    SUM((UnitPrice * Quantity)*(1-Discount)) AS each_price
FROM OrderDetails

GROUP BY OrderID 
ORDER BY each_price ASC
LIMIT 5;
```

```python
%%sql

--What are the 5 largest orders processed
SELECT
    OrderID,
    SUM((UnitPrice * Quantity)*(1-Discount)) AS each_price
FROM OrderDetails

GROUP BY OrderID 
ORDER BY each_price DESC
LIMIT 5;
```
