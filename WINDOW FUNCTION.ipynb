Question: Calculate the land area covered as a percentage of the country's subregion

/* 
    Divide the Land_area by the SUM() BY the areas OVER each Sub_region's PARTITION. Name this column pct_sub_region_land_area.
    Round the calculation off to 4 decimal places.

Add this line to the query from the first task.
*/

```python
%%sql

SELECT
    Sub_region,
    Country_name,
    ROUND(Land_area/SUM(Land_area) OVER(PARTITION BY Sub_region)*100,4) AS pct_sub_region_land_area
FROM 
    Access_to_Basic_Services
WHERE
    Land_area IS NOT NULL
    AND
    Time_period = 2020;   
```

```python
%%sql

SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services
FROM 
    united_nations.Access_to_Basic_Services;
```

Q1.Order countries based on their water access levels per year
Use the ROW_NUMBER() function to order the countries based on their water access levels per year.

```python
%%sql
--calculates the rank of water services based on the Pct_managed_drinking_water_services column within each time period
SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services,
    ROW_NUMBER()OVER(PARTITION BY Time_period
    ORDER BY Pct_managed_drinking_water_services ASC) AS Rank_of_water_access
FROM
    Access_to_Basic_Services;
```

Question: calculates the rank by lowest of water services based on the Pct_managed_drinking_water_services column within each time period

```python
%%sql

SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services,
    ROW_NUMBER()OVER(PARTITION BY Time_period
    ORDER BY Pct_managed_drinking_water_services ASC) AS Rank_of_water_access
FROM
    Access_to_Basic_Services
LIMIT 5;
```

Question: calculates the rank by highest of water services based on the Pct_managed_drinking_water_services column within each time period

```python
%%sql

SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services,
    ROW_NUMBER()OVER(PARTITION BY Time_period
    ORDER BY Pct_managed_drinking_water_services DESC) AS Rank_of_water_access
FROM
    Access_to_Basic_Services
LIMIT 5;
```

```python
%%sql

SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services,
    ROW_NUMBER()OVER(PARTITION BY Time_period
    ORDER BY Pct_managed_drinking_water_services DESC) AS Rank_of_water_access
FROM
    Access_to_Basic_Services;
```

Looking at the year 2016, Korea, Brunei, and Singapore, whose access levels were all 100%, have been ranked differently. 
This is what ROW_NUMBER() does, but we want to rank the countries by access level,
and countries that have the same access level should share the rank

```python
%%sql

SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services,
    ROW_NUMBER()OVER(PARTITION BY Time_period
    ORDER BY Pct_managed_drinking_water_services DESC) AS Rank_of_water_access
FROM
    Access_to_Basic_Services
WHERE
    Pct_managed_drinking_water_services < 90
LIMIT 5;
```

Question: Assess the rankings for countries with the same water level access in a given year

The following query works the same as the previous one with the ROW_NUMBER() function, 
it's just that we now have the RANK() function and
rows with the same values will receive the same rank, and the next rank(s) are skipped accordingly. 

```python
%%sql
SELECT
    Country_name,
    Time_period,
    Pct_managed_drinking_water_services,
    RANK() OVER( PARTITION BY Time_period 
    ORDER BY Pct_managed_drinking_water_services ASC) AS Rank_of_water_services
FROM 
    united_nations.Access_to_Basic_Services;
```

# Using value-based window functions

How to use the `LAG()` value-based window function to extract particular column values from the previous row.
How the results from `LAG()` can be used to perform analysis such as calculating the ratef change between consecutive values

Question: Add a new column with the previous year's percentage of managed drinking water.

```python
%%sql

SELECT
    Country_name,
    Time_period,
    LAG(Pct_managed_drinking_water_services)OVER(PARTITION BY Country_name
    ORDER BY Time_period) AS previous_year_access_to_water
FROM 
    united_nations.Access_to_Basic_Services;
```

```python
%%sql

SELECT
    Country_name,
    Time_period,
    LAG(Pct_managed_drinking_water_services)OVER(
    ORDER BY Time_period) AS previous_year_access_to_water
FROM 
    united_nations.Access_to_Basic_Services;
    
#no partitioning by country_name and order by year(2015,2016...)
```
